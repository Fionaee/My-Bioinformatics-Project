```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# ================= 参数设置 =================
DATA_PATH = r"D:\生信study\Covid19\owid-covid-data.csv"    # 数据文件路径
COUNTRY = "China"         # 指定要分析的国家/地区
# ============================================

# 读取数据
df = pd.read_csv(DATA_PATH)

# 筛选指定国家的数据
country_df = df[df["location"] == COUNTRY].copy()

# 处理日期格式
country_df["date"] = pd.to_datetime(country_df["date"])

# 处理缺失值（用0填充新增病例空白）
country_df["new_cases"] = country_df["new_cases"].fillna(0)

# ================= 绘图设置 =================
fig, ax1 = plt.subplots(figsize=(14, 7))

# 绘制原始每日数据（蓝色柱状图）
ax1.bar(country_df["date"], 
        country_df["new_cases"],
        color="#3498db",
        alpha=0.6,
        label="Daily New Cases")
ax1.set_xlabel('Date', fontsize=12)
ax1.set_ylabel('Daily New Cases', color="#3498db", fontsize=12)
ax1.tick_params(axis='y', labelcolor="#3498db")

# 创建第二个Y轴
ax2 = ax1.twinx()

# 绘制平滑曲线（红色曲线）
ax2.plot(country_df["date"], 
         country_df["new_cases_smoothed"],
         color="#e74c3c",
         linewidth=2,
         label="7-day Average")
ax2.set_ylabel('7-day Average New Cases', color="#e74c3c", fontsize=12)
ax2.tick_params(axis='y', labelcolor="#e74c3c")

# X轴日期格式
ax1.xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m"))
ax1.xaxis.set_major_locator(mdates.MonthLocator(interval=2))
fig.autofmt_xdate()

# 标题和网格
plt.title(f"COVID-19 Cases in {COUNTRY}（2019-2024）", fontsize=16, pad=20)
ax1.grid(alpha=0.3, linestyle="--")
ax1.tick_params(axis='x', rotation=45, labelsize=10)

# 合并图例
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, fontsize=12)

# 自动调整布局
plt.tight_layout()

# 保存高清图
plt.savefig('covid_plot.png', dpi=300, bbox_inches='tight')

# 显示图形
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# 设置全局字体为支持中文的字体
plt.rcParams['font.family'] = 'SimHei'

# ================= 参数设置 =================
DATA_PATH = r"D:\生信study\Covid19\owid-covid-data.csv"    # 数据文件路径
COUNTRY = "China"         # 指定要分析的国家/地区
# ============================================

# 读取数据
df = pd.read_csv(DATA_PATH)

# 筛选指定国家的数据
country_df = df[df["location"] == COUNTRY].copy()

# 处理日期格式
country_df["date"] = pd.to_datetime(country_df["date"])

# 处理缺失值（用0填充新增病例空白）
country_df["new_cases"] = country_df["new_cases"].fillna(0)

# ================= 绘图设置 =================
fig, ax1 = plt.subplots(figsize=(14, 7))

# 绘制原始每日数据（蓝色柱状图），增加柱状图宽度和透明度
bar_width = 0.8
ax1.bar(country_df["date"], 
        country_df["new_cases"],
        color="#3498db",
        alpha=0.8,  # 提高透明度
        width=bar_width,  # 增加柱状图宽度
        label="Daily New Cases")
ax1.set_xlabel('Date', fontsize=12)
ax1.set_ylabel('Daily New Cases', color="#3498db", fontsize=12)
ax1.tick_params(axis='y', labelcolor="#3498db")

# 调整 Y 轴刻度范围和刻度标签数量
ax1.set_ylim(0, country_df["new_cases"].max() * 1.1)  # 适当扩大 Y 轴范围
ax1.locator_params(axis='y', nbins=10)  # 设置 Y 轴刻度标签数量

# 创建第二个Y轴
ax2 = ax1.twinx()

# 绘制平滑曲线（红色曲线）
ax2.plot(country_df["date"], 
         country_df["new_cases_smoothed"],
         color="#e74c3c",
         linewidth=2,
         label="7-day Average")
ax2.set_ylabel('7-day Average New Cases', color="#e74c3c", fontsize=12)
ax2.tick_params(axis='y', labelcolor="#e74c3c")

# 调整第二个 Y 轴刻度范围和刻度标签数量
ax2.set_ylim(0, country_df["new_cases_smoothed"].max() * 1.1)  # 适当扩大 Y 轴范围
ax2.locator_params(axis='y', nbins=10)  # 设置 Y 轴刻度标签数量

# X轴日期格式
ax1.xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m"))
ax1.xaxis.set_major_locator(mdates.MonthLocator(interval=2))
fig.autofmt_xdate()

# 标题和网格
plt.title(f"COVID-19 Cases in {COUNTRY}（2019-2024）", fontsize=16, pad=20)
ax1.grid(alpha=0.3, linestyle="--")
ax1.tick_params(axis='x', rotation=45, labelsize=10)

# 合并图例
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, fontsize=12)

# 自动调整布局
plt.tight_layout()

# 保存高清图
plt.savefig('covid_plot.png', dpi=300, bbox_inches='tight')

# 显示图形
plt.show()
```
